## Identify a metric for this data and calculate it.    
*   order_amount per shop_id  
*   order_amount per payment_method



```python
import pandas as pd

orders_csv = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv", parse_dates=['created_at'])
orders_csv.head(3)
```

```python
orders_df = pd.DataFrame(orders_csv)
orders_df.head(3)
```

```python
shop_df = orders_df[['shop_id', 'order_amount','total_items']]
```

```python
shop_df[['shop_id', 'order_amount', 'total_items']].groupby('shop_id')\
.mean().plot(figsize=(10,6), fontsize=12, title='Order sales per shop')
```

```python
shop_df.groupby('shop_id').agg({'order_amount':sum, 'total_items':sum}).sort_values(['order_amount'],ascending=False)
```

```python
orders_df[['payment_method', 'order_amount', 'total_items']].groupby('payment_method')\
.mean().plot(figsize=(10,6), fontsize=12, title='Order sales per payment method')
```

```python
payment_df = orders_df[['order_amount','total_items','created_at', 'payment_method']]
payment_df.groupby('payment_method').mean('order_amout')
```
